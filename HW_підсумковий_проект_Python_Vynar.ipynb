```python
# Підсумковий проект - Робота з Python, використання бібліотек, очищення, агрегатування, фільтрація даних та створення зрозумілих для читання таблиць.
import pandas as pd
import numpy as np

# Завантаження необхідних датасетів для виконання підсумкового проекту

# Заванатжуєм файл schema.csv на власний googl диск і аналізуємо структуру
# https://drive.google.com/file/d/13LySVF4rYo2LZ34jm0NeL1xtzBL2l9Rd/view?usp=drive_link

file_id = '13LySVF4rYo2LZ34jm0NeL1xtzBL2l9Rd'
url = f'https://drive.google.com/uc?id={file_id}'
df = pd.read_csv(url)

# Заванатжуєм файл results.csv на власний googl диск
# використовуєм функцію gdown через те що файл має великий розмір
# https://drive.google.com/file/d/1JbJNoe_dHkKEYEbJBkbMp7A8m3Hrphu6/view?usp=drive_link

import gdown

file_id_1 = '1JbJNoe_dHkKEYEbJBkbMp7A8m3Hrphu6'
url_1 = f'https://drive.google.com/uc?id={file_id_1}'
gdown.download(url_1, 'results.csv', quiet=False)

df_1 = pd.read_csv('results.csv')
```

**Завдання 2. Аналіз повноти відповідей респондентів**

```python
# Переглядаємо структуру та інформацію про файл "schema.csv"
df.info()
print('\n==============================================================================================================')
df.head(3)
```

```python
# Створюємо множину унікальних питань, які є в списку та в колонках df
# унікальні ID блоків питань у датасеті
q_name = df['qname'].unique()
relevant_questions = set(np.unique(q_name)) & set(df_1.columns)
relevant_questions

# Видаляємо рядки, де є пропуски (NaN) у цих питаннях
df_1_complete = df_1.dropna(subset=relevant_questions)

# Рахуємо кількість респондентів, які відповіли на ВСІ питання (shape[0] - кількість рядків)
full_responses_count = df_1_complete.shape[0]

print('=============================================================')
print(f'Кількість респондентів, які відповіли на всі питання: {full_responses_count}')
```

**Завдання 3. Статистичний аналіз досвіду респондентів**

```python
# Готуємо колонку 'WorkExp' для аналізу даних, перевіряємо на наявність пропущених значень чи помилок
# df_1['WorkExp'].dtype
# df_1['WorkExp'].describe()
# df_1['WorkExp'].isna().sum() # отримуємо результат np.int64(6298) к-сті клітинок з значенням NaN
# Очищаємо колонку від відсутніх значень NaN
work_exp = df_1['WorkExp'].dropna()

print(f'Міри центральнгої тенденції : ')
print('=============================')

# Обчислюємо середнє значення (mean)
print(f'mean :  {round(work_exp.mean(), 1)}')
print('=============================')

# Обчислюємо медіану (median)
print(f'median:  {round(work_exp.median(), 1)}')
print('=============================')

# Обчислюємо моду (mode)
print(f'mmode:  {round(work_exp.mode(), 1).tolist()}')
```

**Завдання 5. Визначення популярності Python**

```python
# Визначаєм к-сть згадувань Python в колонці 'LanguageHaveWorkedWith' та його відсотка популярності

# Залишаємо рядки які відповідають умові де хочаб один раз згадується Python
worked_with_python = df_1['LanguageHaveWorkedWith'].str.contains(
    'Python',
    case=False,
    na=False
)
# Визначаєм к-сть респондентів що працюють з Python
python_users = worked_with_python.sum()

# Визначаєм загальну к-сть респондентів
total_users = len(df_1)

# Розраховуємо відсоток таких респондентів та виводим результат
python_percent = (python_users / total_users)*100

print('=============================')
print("Відсоток респондентів:", round(python_percent, 1), "%")
```

**Завдання 6. Аналіз шляхів навчання програмуванню**

```python
# Проглядаємо дані стовпчика 'LearnCode' для визначення мітки для фільтрації і відбору даних
df_1['LearnCode'].dropna().unique() # маємо колонку з списками у яких багато слів

# Сворюєм додатковий стовпець для підрахунку к-сті респондентів які навчались на онлайн курсах
df_1['learned_online_courses'] = df_1['LearnCode'].str.contains(
    'Online',
    case=False,
    na=False
)
# Визначаєм відсоток користувачів які обрали навчання на онлайн курсах та виводим результат
online_courses_count = df_1['learned_online_courses'].sum()
total_users = len(df_1)
online_percent = (online_courses_count / total_users) * 100

print('=============================')
print("Відсоток респондентів:", round(online_percent, 1), "%")
```

**Завдання 7. Географічний аналіз компенсації Python-розробників**

```python
# Створюємо датафрейм з працівниками які працювали з Python
df_worked_with_python = df_1[df_1['LanguageHaveWorkedWith'].str.contains(
    'Python',
    case=False,
    na=False
)]

# Видаляємо пропущені значення компенсації в колонці 'ConvertedCompYearly'
df_worked_with_python = df_worked_with_python[df_worked_with_python['ConvertedCompYearly'].notna()]

# Групування по країнах
grouped_country = df_worked_with_python.groupby('Country')

# Розрахунок 'mean', 'median'
contry_mean_median= grouped_country['ConvertedCompYearly'].agg(['mean', 'median']).round()

print(' Таблиця середнього значення і медіани зарплати респондентів що працюють з Python в розрізі країн ')
print('======================================================================================================')
print(contry_mean_median)
```

**Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів**

```python
# Створюємо новий датафрейм і вибираємо потрібні колонки
df_new = df_1[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]

# Видаляємо рядки з пропущеною зарплатою
df_new = df_new.dropna(subset=['ConvertedCompYearly'])

# Сортуємо за зарплатою у спадному порядку
df_new = df_new.sort_values(by='ConvertedCompYearly', ascending=False)

# Беремо топ-5 респондентів по зарплаті
top_5 = df_new.reset_index(drop=True).head(5)

# Виводимо результат
print('===================================================================')
print(f'Топ-5 респондентів з найвищою зарплатою : {top_5}')
```
